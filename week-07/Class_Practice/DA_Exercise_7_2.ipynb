This notebook analyzes employee sales performance for RetailHub Inc. It uses Pandas to create, inspect, filter, sort, and summarize employee sales data.

## 1. Import Pandas and Create the DataFrame

```python
import pandas as pd 

# Employee dictionary
data = {
    'name': ['Alice', 'Ben', 'Carmen', 'Dev', 'Eve', 'Frank', 'Grace', 'Hiro'],
    'city': ['London', 'Manchester', 'London', 'Leeds', 'Birmingham', 'Manchester', 'Leeds', 'London'],
    'sales': [4200.0, 3100.0, 5050.0, 2800.0, 1950.0, 3750.0, 4600.0, 3300.0],
    'months': [12, 6, 24, 3, 18, 9, 15, 7]
}

# Turning the dictionary into a Pandas DataFrame, which is like a table
df = pd.DataFrame(data)

# Displays full table
print(df)
print()
# Shows how many rows and columns
print(df.shape)
print()
# Shows the data type of each column
print(df.dtypes)
print()
# Gives a more in depth summary, including column names, non-null values and data types
df.info()
```

## 2. Inspect and Describe the Data

```python
print("First 3 rows:")
print(df.head(3))
# Head() means to show the top rows, if rows aren't specified it will default to 5 rows, req is asking 3 rows

print("\nLast 3 rows:")
print(df.tail(3))
# This will show the bottom rows, specifically the last 3 employees

print("\nSummary Statistics:")
print(df.describe())
# describe() is a pandas method, it will calculate numeric columns automatically
# My output shows the count, mean, std, min, 25%, 50%, 75% and then max

# This is getting the sales column and calculating the average
print(f"\nMean sales: £{df['sales'].mean():,.2f}")
# This is getting the sales column again and finding the highest value
print(f"Max sales: £{df['sales'].max():,.2f}")
```

## 3. Select, Filter and Sort

```python
# Select only name and sales columns
# 2 brackets were used to give multiple columns
print("Name and Sales Columns:")
print(df[['name', 'sales']])

# Filter employees with sales greater than 3500
high_sales = df[df['sales'] > 3500.0]

print("\nEmployees with Sales > £3500:")
print(high_sales)

# Sort sales from highest to lowest
df_sorted = df.sort_values('sales', ascending=False)

print("\nSorted by Sales (Highest First):")
print(df_sorted)

# Show the top-selling employee by using [0] in the sorted list
print("\nTop Seller Record:")
print(df_sorted.iloc[0])
```

## 4. Commission Column

```python
# Add commission column
# Get the sales values and multiply by 8%
df['commission'] = df['sales'] * 0.08

# Add rating column using lambda and apply()
# .apply() will go through every row and apply the logic
# lambda x: will take one sales value at a time and see if the first condition is met
# If it failed, it will test for the second condition and finally if both failed it will print out 'Developing'
df['rating'] = df['sales'].apply(
    lambda x: 'Star' if x >= 4000
    else ('Good' if x >= 2500 else 'Developing')
)

# Print updated DataFrame
print("Updated DataFrame:")
print(df)

# Print commission per employee
print("\nCommission Per Employee:")

for _, row in df.iterrows():
    print(f"{row['name']} | £{row['commission']:,.2f} | {row['rating']}")
```

## 5. GroupBy — Sales by City

```python
# Total sales by city
city_totals = df.groupby('city')['sales'].sum()

print("Total Sales by City:")
print(city_totals)

# Mean sales by city
city_means = df.groupby('city')['sales'].mean()

print("\nMean Sales by City:")
print(city_means)

# Full city summary
city_summary = df.groupby('city').agg({
    'sales': ['sum', 'mean', 'count']
})

print("\nFull City Summary:")
print(city_summary)

# Find top city
top_city = city_totals.idxmax()

print(f"\nTop City: {top_city}")
print(f"Total Sales: £{city_totals[top_city]:,.2f}")
```

## Summary

London is the top-performing city with the highest total sales of £12,550.00.
Three employees achieved a Star rating, while Eve was rated Developing.
Overall, employee sales performance shows strong results, especially in London and Leeds.
