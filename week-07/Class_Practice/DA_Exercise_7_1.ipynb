This notebook uses NumPy to analyze the weekly revenue for five regional offices. It calculates statistics, applies bonus, filter above-average regions, and reshapes the dtaa into 2D arrays.

```python
import numpy as np

# Creating the 1D array revenue to store the five weekly revenue values
revenue = np.array([52400.0, 38750.0, 61200.0, 47900.0, 55650.0])

print(f"Array : {revenue}")         # Shows the actual revenue values
print(f"dtype : {revenue.dtype}")   # Shows the data type | I think for whole numbers it will be 32 and for float it will be 64 
print(f"shape : {revenue.shape}")   # Shows the structure of the array or how much values are in it
```

```python
# Using Numpy's built in functions for the calculation
print("----- Revenue Statistics -----")
print(f"Total   : ${np.sum(revenue):,.2f}")
print(f"Mean    : ${np.mean(revenue):,.2f}")
print(f"Median  : ${np.median(revenue):,.2f}")
print(f"Std Dev : ${np.std(revenue):,.2f}")
print(f"Minimum : ${np.min(revenue):,.2f}")
print(f"Maximum : ${np.max(revenue):,.2f}")
```

```python
# Calculation for the bonus | Element-wise Arithmetic
# revenue * 1.05 means every number gets multiplied by 1.05
# The 1 keeps the original amount, and the .05 adds the extra 5%
revenue_bonus = revenue * 1.05
bonus_amount = revenue_bonus - revenue

print("Original Revenue   :", revenue)
print("Revenue + 5% Bonus :", revenue_bonus)
print("Bonus Amount       :", bonus_amount)
```

```python
# This creates a true of false check for each region
# If the region's revenue is above the average, it will become true
# Then revenue[above_avg] only grabs the values where the answer was true
above_avg = revenue > np.mean(revenue)

print("Mask True/False       :", above_avg)
print("Above average         :", revenue[above_avg])
print(f"Regions above average : {np.sum(above_avg)}")
```

```python
# reshape(5, 1) changes the revenue array into 5 rows and 1 column
# np.vstack() stacks the arrays so there's 1 row for the original revenue and another row fror the bonus revenue
# axis=1 works across each row so it gives row totals
# axis=0 works down each column so it gives the average for each region
col_vec = revenue.reshape(5,1)
print("Column vector (5,1):\n", col_vec)

q_data = np.vstack([revenue, revenue * 1.05])
print("\n2D array:\n", q_data)

print("\nWeekly totals  :", np.sum(q_data, axis=1))
print("Region average :", np.mean(q_data, axis=0))
```

## Summary

The five regions generated a total of $255,900.00 in original weekly revenue. Three out of five regions performed above the average revenue of $51,180.00. After applying a 5% bonus, the total revenue increased to $268,695.00.
