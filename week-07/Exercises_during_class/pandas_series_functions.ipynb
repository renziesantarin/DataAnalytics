## Pandas Series Funtion Exercise

This activity will create and explore a Pandas series using numeric values: 10, 20, 30, 20, NaN.

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Creating Pandas Series
pandas_series = pd.Series([10, 20, 30, 20, np.nan])

# Displaying original series
print("Original Series:")
print(pandas_series)
```

```python
# Statistical functions
print("\nSum:", pandas_series.sum())
print("Mean:", pandas_series.mean())
print("Maximum Value:", pandas_series.max())
print("Minimum Value:", pandas_series.min())
```

```python
# Display first three values
print("\nFirst 3 Values:")
print(pandas_series.head(3))
```

```python
# Check for missing values
print("\nMissing Values:")
print(pandas_series.isnull())
```

```python
# Replace missing values with 0
filled_series = pandas_series.fillna(0)

print("\nSeries with Missing Values Replaced:")
print(filled_series)
```

```python
# Sort values
sorted_series = pandas_series.sort_values()

print("\nSorted Series:")
print(sorted_series)

# Count repeated values
print("\nValue Counts:")
print(pandas_series.value_counts())
```
