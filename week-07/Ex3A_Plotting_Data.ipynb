# Exercise 3.A Getting familiar with Matplotlib

Start by importing numpy (as np) AND matplotlib.pyplot (as plt) into the notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

Create two variables.

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

print(x)
print(y)

# The x list will represent the values on the x-axis (the horizontal bottom part of the graph), and the y list will 
# represent the values on the y-axis (the vertical side of the graph). 
# I added print() just so I can quickly confirm the lists were created correctly before graphing.
```

Use plt.plot() to plot the line graph for x and y, and then display it using plt.show().

```python
plt.plot(x, y)

plt.show()

# Creating a graph using the x and y values I made earlier. The x list controls where the points go across the bottom (horizontal axis), 
# and then the y list controls how high or low the points appear.
```

Duplicate the previous code cell, then update it to add markers to. the data points using the marker='o' parameter.

```python
plt.plot(
    x, y, 
    marker='o'
)

plt.show()

# Taking the same graph and adding markers to make the individual data points easier to see. The marker='o' puts a small circle on every data point.
# So instead of only seeing the line, I can now clearly see where each actual data value is located on the graph.
```

Duplicate the previous code cell, then customize the plot with a title (‘Example Graph’), x-axis label (‘X axis’), and y-axis label (‘Y axis’).

```python
plt.plot(
    x, y, 
    marker='o'
)

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()

# Customizing the graph to make it easier to understand. I’m still plotting the same x and y values with markers, 
# but now I’m adding a title and labels for x and y-axis.
```

## Styling Plots

Plot the same data as above, but this time, change the line style to dashed ('--') and the line color to red.

```python
plt.plot(
    x, y, 
    marker='o', 
    linestyle='--', 
    color='red'
)

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()

# Still plotting the same x and y values, but this time changing the lines to be dashed instead of solid.
# And then changing the color from blue to red.
```

Change the marker style to a square ('s'), set the marker size to 10, and change the marker color to green.

```python
plt.plot(
    x, y,
    linestyle='--',
    color='red',
    marker='s',
    markersize=10,
    markerfacecolor='green'
)

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()

# Still using the same x and y data, and I’m keeping the dashed red line, but now instead of circles, I'm going to use square,
# specifying the marker size to 10 and changing the marker color to green.
```

Experiment with at least two different additional marker styles and colors, and display the plots.

```python
# Example 1:
plt.plot(
    x, y,
    linestyle='-',
    color='blue',
    marker='^',
    markersize=10,
    markerfacecolor='orange'
)

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()

# Changed the line back to a solid line ('-') and made it blue, while the triangle markers are orange.
```

```python
# Example 2:
plt.plot(
    x, y,
    linestyle=':',
    color='purple',
    marker='D',
    markersize=10,
    markerfacecolor='yellow'
)

plt.title("Example Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()

# Changed the line style to dotted (':') and made the line purple, while the diamond markers are yellow.
```

## Advanced Plotting Techniques

Create a bar chart showing the sales of different products.

```python
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [23, 17, 35, 29]

plt.bar(products, sales)
plt.show()

# Starting by creating two lists. The products list holds the product names that will appear on the x-axis, 
# and the sales list holds the sales numbers that will determine the height of each bar. Then I pass both lists into plt.bar().
```

Customize your bar chart by adding a title, x-axis label, and y-axis label. Set the color of the bars to 'skyblue' and adjust the bar width to 0.5.

```python
plt.bar(
    products, 
    sales, 
    color="skyblue", 
    width=0.5
)

plt.title("Sales by Product")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.show()

# I pass products and sales into plt.bar() again, but this time I add color="skyblue" to make the bars sky blue 
# and width=0.5 to make the bars a little narrower. Then I add a title so the viewer knows what the chart is about, 
# an x-axis label to show that the bottom categories are products, and a y-axis label to show that the bar heights represent sales.
```

Create a scatter plot to visualize the relationship between two variables.

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()

# Creating two new variables, x and y, using np.random.rand(50). This tells NumPy to generate 50 random values 
# between 0 and 1 for each variable. The x values control where the points go horizontally, and the y values control 
# where the points go vertically. Then I pass both variables into plt.scatter(), which creates a scatter plot.
```

Customize your scatter plot by setting the size of the markers based on another variable.

```python
x = np.random.rand(50)
y = np.random.rand(50)
s = np.random.rand(50) * 100

plt.scatter(x, y, s=s, c=s, cmap="viridis")

plt.title("Random Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()

# I create x and y again as 50 random values, and then I create a third variable called s. The s variable controls the size of each marker. 
# Since np.random.rand(50) creates small decimal values between 0 and 1, I multiply it by 100 to make the marker sizes easier to see.
# Then I pass x, y, and s into plt.scatter(). The s=s part tells Matplotlib to use my s values as the marker sizes. 
# The c=s part tells Matplotlib to also use those same values to control the marker colors. 
# The cmap="viridis" part tells Matplotlib which color map to use.
```

## Subplots

Create a figure with two subplots: one for a line chart and one for a histogram. 
In the first subplot, plot a sine wave using x.
In the second subplot, create a histogram of 1000 random numbers generated using np.random.randn(1000).
Add titles and labels to each subplot.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# FIRST SUBPLOT: SINE WAVE
x = np.linspace(0, 10, 100)
y = np.sin(x)

axes[0].plot(x, y)
axes[0].set_title("Sine Wave")
axes[0].set_xlabel("X Values")
axes[0].set_ylabel("Sine of X")

# SECOND SUBPLOT: HISTORGRAM
random_numbers = np.random.randn(1000)

axes[1].hist(random_numbers)
axes[1].set_title("Histogram of Random Numbers")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

plt.show()

# I created two chart spaces side by side so I could compare different visualizations in one figure. 
# In the first chart, I generated x-values and used them to create a smooth sine wave, which shows a repeating up-and-down pattern. 
# In the second chart, I generated 1,000 random numbers and displayed them as a histogram to show how the values are distributed. 
# Finally, I added titles and labels to make both charts easier to read and understand.
```
