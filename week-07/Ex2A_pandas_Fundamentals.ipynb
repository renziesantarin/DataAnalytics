## Series

```python
# Creating a list object
aa = [1, 2, 3, 4]

# Creating a tuple object
ab = (5, 6, 7, 8)

# Creating a set object
ac = {9, 10, 11, 12}

# Creating a dictionary object
ad = {
    'first': 13,
    'second': 14,
    'third': 15,
    'fourth': 16
}
```

```python
# Creating a Series from list aa
ser1 = pd.Series(aa)

# Creating a Series from tuple ab
ser2 = pd.Series(ab)

# Creating a Series from set ac
# Initially did ser3=pd.Series(ac) but it gave me a TypeError since sets are unordered
# Converting set to list
ser3 = pd.Series(list(ac))

# Creating a Series from dictionary ad
ser4 = pd.Series(ad)

# Printing each Series
print(ser1)
print()

print(ser2)
print()

print(ser3)
print()

print(ser4)
```

```python
# Creating a Series using literal values
ser5 = pd.Series(17, 18, 19, 20)

# Display the Series
print(ser5)

# TypeError: Index(...) must be called with a collection of some kind, 18 was passed
# The problem is that pandas thought 18 was supposed to be the index, but it was not given as a list or collection.
```

## DataFrames from Series

```python
# Creating a variable df1 and storing a DataFrame in it
# Using data from ser1 to create a table
df1 = pd.DataFrame(ser1)

# Displaying df1
df1
```

```python
# Inserting ser2 into column position 1
df1.insert(1, 'Column_2', ser2)

# Inserting ser3 into column position 2
df1.insert(2, 'Column_3', ser3)

# Display updated DataFrame
df1

# insert(index position, column name, data)
```

```python
# Adding ser4 as a new column
df1['Column_4'] = ser4

# Display the updated DataFrame
df1
```

ser4 behaved differently because it was created from a dictionary, which uses custom index labels ('first', 'second', 'third', 'fourth') instead of numbered indexes (0,1,2,3).

Since df1 uses numbered indexes and ser4 uses text labels, pandas could not match the rows correctly. Because of this mismatch, pandas filled the new column with NaN values.

```python
# .loc is creating row 4 and adding the specified values inside
df1.loc[4] = [5, 9, 12, 0]

# Display updated DataFrame
df1
```

- The dataset contains 262 records for every column, confirming there are no missing values.

- The language column contains only 1 unique value, meaning all movies are in English.

### Less useful summarizations

- Since the dataset contains mostly text/string data, numerical statistics like averages or totals are not available.
- The summary is useful for identifying repeated values and categories, but not useful for mathematical analysis.

```python
# Reading northwind_example.csv and turning it into a DataFrame
df3 = pd.read_csv('northwind_example.csv')

# Displaying the DataFrame
df3
```

```python
# Displaying column statistics
df3.describe()
```

### Insights from describe() for df3

- The dataset contains 500 records for all numeric columns, meaning there are no missing numeric values.
- The average (mean) UnitPrice is approximately 23.37, while the average Quantity ordered is about 24 items.
- Product prices vary significantly, ranging from 2.00 to 210.80.
- Order quantities range from 1 item to 120 items, showing that some orders are much larger than others.
- The median (50%) quantity ordered is 20, meaning half of the orders were above 20 items, and half were below.

### How this summary differs from the previous example

- This dataset contains numeric data, so describe()` provides mathematical statistics such as mean, standard deviation, minimum, maximum, and percentiles.
- The previous dataset (df2) contained text/string data, so describe() focused on unique values, most common values (top), and frequency (freq) instead.
- Numeric datasets allow for deeper mathematical analysis, while text datasets are better for category and frequency analysis.
