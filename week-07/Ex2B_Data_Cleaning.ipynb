## Filtering a DataFrame

Filter the DataFrame to show only rows where Age is greater than 18.

```python
df[df["Age"] > 18]

# Focusing on the Age column and checking which ages are greater than 18. Then I pass that condition 
# back into the DataFrame so pandas only shows the rows where that condition is true.
```

Filter the DataFrame to show only rows where Score is greater than 50 and less than 100.

```python
df[(df["Score"] > 50) & (df["Score"] < 100)]

# Focusing on the Score column and checking two things at the same time. First, the score has to be greater than 50. 
# Second, the score has to be less than 100. The & means both conditions must be true.
```

## Modifying a DataFrame

Create new variable df_cleaned. Assign it the value of df, using the drop function to remove rows where the Score is less than 0 
(i.e., invalid score data). Then call df_cleaned to review the result.

```python
df_cleaned = df.drop(df[df["Score"] < 0].index)

df_cleaned

# Focusing on the Score column and checking two things at the same time. First, the score has to 
# be greater than 50. Second, the score has to be less than 100. The & means both conditions must be true.
# I save the result as df_cleaned so my original df stays safe, and then I display df_cleaned to check that the invalid row was removed.
```

Continuing to work with df_cleaned for the rest of these steps, remove the City column from the DataFrame using drop. Call df_cleaned to review the result.

```python
df_cleaned = df_cleaned.drop(columns=["City"])

df_cleaned

# I pass the City column into drop() and tell pandas I want to remove that column. 
# Then I save the updated table back into df_cleaned and display it to confirm that City is gone.
```

Replace all missing Age values with the mean Age of the DataFrame using fillna. Call df_cleaned to review the result.

```python
df_cleaned["Age"] = df_cleaned["Age"].fillna(df_cleaned["Age"].mean())

df_cleaned

# Focusing on the Age column because some ages are missing. First, I calculate the average age using .mean(). 
# Then I use fillna() to replace the missing Age values with that average. Finally, I save the updated Age column back into df_cleaned.
```

Replace any Score value of 0 with the mean Score of the DataFrame using replace. Call df_cleaned to review the result.

```python
df_cleaned["Score"] = df_cleaned["Score"].replace(0, df_cleaned["Score"].mean())

df_cleaned

# Focusing on the Score column again. This time, I’m looking for scores that are exactly 0. Since the exercise treats 0 as 
# something that should be replaced, I calculate the average score and use that average to replace the 0 values.
```

Remove any remaining rows that contain NaN values using the dropna function. Call df_cleaned to review the result.

```python
df_cleaned = df_cleaned.dropna()

df_cleaned

# Checking the whole DataFrame for any remaining missing values. If a row still has a NaN, that row gets removed.
# Then I save the cleaned version back into df_cleaned.
```

Apply the values from ID as index labels, then remove column ID. (Pay attention to the numbering!) Call df_cleaned to review the result.

```python
df_cleaned = df_cleaned.set_index("ID")

df_cleaned

# Taking the ID column and making it the index, which means it becomes the row label on the left side
# instead of staying as a normal column. Then I display the final cleaned DataFrame.
```
