# FinTrack - Personal Budget Tracker
**Course:** Data Analytics | **Date:** 5/11/2026

This notebook collects monthly income and expenses, then produces a formatted budget summary - no libraries required.

## Section 1: Analyst Information
Enter your name and the month being analysed when prompted.

```python
analyst = input("Analyst name: ")
month = input("Budget month: ")
```

## Section 2: Monthly income and expenses
Enter each dollar amount when prompted. Values are cast to float so arithmetic works correctly - no libraries needed.

```python
income = float(input("Monthly income $: ")) 
rent = float(input("Rent / housing $: ")) 
food = float(input("Food / groceries $: ")) 
transport = float(input("Transport $: ")) 
utilities = float(input("Utilities $: "))
entertainment = float(input("Entertainment $: "))
```

## Section 3: Budget calculations
Total expenses, remaining budget, and savings rate are computed using plain Python arithmetic - no import statements required.

```python
total_expenses = rent + food + transport + utilities + entertainment
remaining = income - total_expenses 
savings_rate = (remaining / income) * 100
```

## Section 4: Budget summary report
The report below is generate using Python f-strings.

```python
print(f"""
====== FinTrack Budget Report ======
Analyst: {analyst}
Month: {month}
------------------------------------
Income         : ${income:,.2f}
Rent/Housing   : ${rent:,.2f}
Food/Groceries : ${food:,.2f}
Transport      : ${transport:,.2f}
Utilities      : ${utilities:,.2f}
Entertainment  : ${entertainment:,.2f}
------------------------------------
Total expenses : ${total_expenses:,.2f}
Remaining      : ${remaining:,.2f}
Savings rate   : {savings_rate:.1f}%
====================================
""")
```

```python
if savings_rate >= 20:
    print("Healthy")
elif savings_rate >=10:
    print("Tight")
else:
    print("Over budget")
```
