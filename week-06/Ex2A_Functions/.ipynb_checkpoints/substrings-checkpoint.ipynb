## Lab 1. Working with names - parsing strings

```python
test_name = 'John Doe'
```

```python
def parse_and_display(test_name):
    
    for i in range(len(test_name)): # Why does it have to be i? It doesn't have to be, but it is usually used because it stands for index or iterator.
        
        if test_name[i] == ' ':
            first = test_name[:i]
            last = test_name[i+1:]  # Adding +1 here to exclude the space before last name.
            
    print(f"Full name  : {test_name}")
    print(f"First name : {first}")
    print(f"Last name  : {last}")
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```python
parse_and_display(test_list[0])
```

```python
parse_and_display(test_list[1])
```

```python
parse_and_display(test_list[2])
```

```python
parse_and_display('Marenza Santarin')
```

```python
def parse_and_display2(name):

    # For a one-word name
    if name.count(' ') == 0:
        first = name
        print(f"First name: {first}")

    # For first and last name
    elif name.count(' ') == 1:
        for i in range(len(name)):     # Checking every character in the name
            if name[i] == ' ':         # Finding the space in the name
                first = name[:i]       # Everything before the space is the first name
                last = name[i+1:]      # Everything after the space is the last name, + 1 to skip the space
        print(f"First name: {first}")
        print(f"Last name: {last}")

    # For first, middle, and last name
    elif name.count(' ') == 2:         # Checking for 2 spaces
        spaces = []                    # Because slicing uses index positions, the spaces variable is created to store space locations

        for i in range(len(name)):
            if name[i] == ' ':         # This will loop through every letter and find a space and then save its position       
                spaces.append(i)       # Adding the space location to the list, from [] to [8, 12]

        first = name[:spaces[0]]                # Getting everything before the first space
        middle = name[spaces[0]+1:spaces[1]]    # Getting everything between the first and second space
        last = name[spaces[1]+1:]               # Getting everything after the second space, + 1 to skip the space

        print(f"First name: {first}")
        print(f"Middle name: {middle}")
        print(f"Last name: {last}")
```

```python
parse_and_display2('MaoMao')
```

```python
parse_and_display2('King Bumi')
```

```python
parse_and_display2('Frieren The Slayer')
```
