# Exercise 2.A Using built-in functions

## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(part_code):              # Creating a function to get product code
    colon_position = part_code.index(':')      # This is finding where the colon is located and stores that index number
    supplier = part_code[:colon_position]      # This is saying everything from the beginning up to BUT not including the colon is the supplier
    return supplier                            # This will send the supplier code back so it can be used later in a print statement
```

```python
get_supplier_code('ACME:123-L') # Testing
```

```python
def get_product_num(part_code):                               # Creating a function to get product number
    colon_position = part_code.index(':')                     # This is finding where the colon is located and stores that index number
    hyphen_position = part_code.index('-')                    # This is finding where the hyphen is located and stores that index number
    product_num = part_code[colon_position+1:hyphen_position] # This is saying everything after the colon+1 BUT before the hyphen is the product number
    return product_num                                        # This will send the product number back so it can be used later in a print statement
```

```python
get_product_num('ACME:123-L') # Testing
```

```python
def get_size(part_code):                       # Creating a function to get the size
    hyphen_position = part_code.index('-')     # This is finding where the hyphen is located and stores that index number
    size = part_code[hyphen_position+1:]       # This is saying everything after the hyphen+1 is the size
    return size                                # This will send the size back so it can be used later in a print statement
```

```python
get_size('ACME:123-L') # Testing
```

```python
acme_part = 'ACME:123-L'
di_part = 'DI:12345-M'
ace_part = 'ACE:1-12'
```

```python
# For ACME
print(f"For {acme_part}, the supplier is {get_supplier_code(acme_part)}.")
print(f"Part number {get_product_num(acme_part)} is available in size {get_size(acme_part)}.")
```

```python
# For DI
print(f"For {di_part}, the supplier is {get_supplier_code(di_part)}.")
print(f"Part number {get_product_num(di_part)} is available in size {get_size(di_part)}.")
```

```python
# For ACE
print(f"For {ace_part}, the supplier is {get_supplier_code(ace_part)}.")
print(f"Part number {get_product_num(ace_part)} is available in size {get_size(ace_part)}.")
```
