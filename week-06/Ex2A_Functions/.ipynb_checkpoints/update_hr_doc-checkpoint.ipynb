## Lab 3. Employee data

This lab exercise makes updates to information extracted from
an employee database.

```python
hr_list = [
('0123', 'HR', 'Rebecca Yang', '69000'),
('0121', 'IT', 'Mark Blick', '67000'),
('0068', 'IT', 'Kahna Larsen', '112000'),
('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
# Update Mark's name
hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')
# Going to the second employee record, which is index 1, then replacing it
# with the new tuple (the whole thing), since tuples cannot be edited after creation.
```

```python
print(hr_list)
```

```python
# Update Jim's department and salary
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
# Going to the second employee record, which is index 1, then replacing it
# with the new tuple (the whole thing), since tuples cannot be edited after creation.
```

```python
print(hr_list)
```

```python
# Display contents in the specific format from the lab workbook
for employee in hr_list:
    print(f"{employee[0]} | {employee[1]} | {employee[2]} | {int(employee[3]):,}")
# Looping one at a time through hr_list
# Printing everything using indexes, converted salary to an integer, and added a comma
```
