# Define a function with parameters and no parameters

```python
def greeting():
    name = input("Please enter your name: ")
    return name

result = greeting() 
print(f"Hello, {result}!")
```

```python
def greeting(name, city, hobby):
    return name, city, hobby

result = greeting(
    input("Please enter your name: "),
    input("Please enter your city: "),
    input("Please enter your hobby: ")
)
print(type(result))

print(f"Hello, {result[0]}! You are from {result[1]} and you enjoy {result[2]}.")
```

```python
def student_info(name, major):                         # Creating or defining the function student_info
    message = f"Hello {name}, your major is {major}."  # Stores a formatted sentence using an f-string
    return message                                     # Instead of printing inside the function, I am sending the completed sentence back to the main program

result = student_info("MaoMao", "Biology")             # The returned value gets stored here

print(result)                                          # Printing whatever is stored inside result
```

# Demonstrating features of strings as a sequence

```python
# 1. Creating a string
text = "Python"
```

```python
# 2. Indexing
print("First character", text[0])
print("Last character", text[-1])
```

```python
# 3. Slicing
print("First three characters:", text[0:3])   # Start at index 0 and stop before index 3
print("From index 2 to end:", text[2:])       # Start at position 2 and keep going
print("Every second character:", text[::2])   # Taking every second character [start, stop, step]
```

```python
# 4. Iteration
print("Characters in string:")
for char in text:
    print(char, end=" ")          # Instead of going to a new line, it's adding a space
print()
```

```python
# 5. Membership
print("'Py' in text?", "Py" in text)
print("'Java' in text?", "Java" in text)
```

```python
# 6 Length of string
print("Length of string:", len(text))
```

```python
# 7. Concatenation and repetition
print("Concatenation:", text + "3.1")
print("Repetition:", text * 2)
```

```python
# 8. Immutability demonstration
try:
    text[0] = "J"
except TypeError as e:
    print("Strings are immutable:", e)
```

# Splitting a string

```python
def parse_code(product_code):                             # Creating a function called parse_code with parameter product_code that will store the code
    parts = product_code.split("-")                       # Splitting the string whenever there's a hyphen
    company = parts[0]                                    # The company variable will store the first one which is in index 0
    product_id = parts[1]                                 # The product_id will store the second one which is in index 1
    category = parts[2]                                   # The category will store the third one which is in index 2
    return company, product_id, category                  # Sending all 3 values back out of the function

company, product_id, category = parse_code("ACME-123-L")  # Calling the function and passing "ACME-123-L"

print(f"Company: {company}")                              # Printing each separately
print(f"Product ID: {product_id}")
print(f"Category: {category}")
```

# Range

```python
# 1. Basic range
print("range(5):")
for i in range(5):
    print(i, end=" ")
print()
```

```python
# 2. Range with start and stop
print("\nrange(2, 8):")
for i in range(2,8):
    print(i, end=" ")
print()
```

```python
# 3. Range with start, stop and step
print("\nrange(0, 20, 5):")
for i in range (0, 20, 5):
    print (i, end=" ")
print()
```

```python
# 4. Counting backwards using a negative step
print("\nrange(10, 0, -2):")
for i in range(10, 0, -2):
    print(i, end=" ")
print()
```
