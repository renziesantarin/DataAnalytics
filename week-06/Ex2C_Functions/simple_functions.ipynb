This is Lab 1 of Exercise 2.C. In this lab, I will be working with custom functions in Python.

```python
# Creating a function 'favorite_things' that has 3 parameters and should have an output of 3 lines
def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician/band is {music}."
print(favorite_things("Renzie", "The Avengers", "Ben&Ben"))
# Calling the function and replacing the parameters with values.
```

```python
# Creating a function 'why_im_here' that has 1 parameter 'purpose' and it will store my answer on why I joined YUU Data Analyst program
def why_im_here(purpose):
    return purpose
print(why_im_here("I joined YUU Data Analytics program to build my technical skills and gain hands-on experience to help me prepare with my career transition to data analytics."))
# Calling the function and added the temporary sentence
```

```python
# Creating a function 'favorite_place' with 2 parameters that will store your favorite place to visit and why
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
print(favorite_place("Stanserhorn, Switzerland", "I love the scenery and the food"))
# Calling the function and added 
```

```python
# Testing and providing a new place and why
print(favorite_place("Imus, Cavite", "it's my hometown and where most of my family and friends reside"))
```
