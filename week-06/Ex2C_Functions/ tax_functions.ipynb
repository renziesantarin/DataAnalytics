# Lab 3 Tax Functions
This lab shows how to define custom Python functions to calculate Social Security tax, Medicare tax, and federal tax withholding based on gross pay and withholding code.

```python
# Creating the first function that accepts 1 piece of info
# gross_pay gets multiplied by 6.2% (tax rate), and then the answer gets returned out of the function
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

```python
# Creating the first function that accepts 1 piece of info
# gross_pay gets multiplied by 1.45% (tax rate), and then the answer gets returned out of the function
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

```python
# Creating the first function that accepts 2 pieces of info
# Using if/elif/else to decide which tax rate to use
# Withholding code 4 starts at 18%, then every number above 4 reduces the tax rate by 0.5%
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        tax_rate = 0.23
    elif withholding_code == 1:
        tax_rate = 0.21
    elif withholding_code == 2:
        tax_rate = 0.195
    elif withholding_code == 3:
        tax_rate = 0.185
    else:
        tax_rate = 0.18 - ((withholding_code - 4) * 0.005)

    return gross_pay * tax_rate
```

```python
# Person 1:
gross_pay = 750
withholding_code = 0

print("Person 1")
print("Social Security Tax: $", round(get_soc_sec_tax(gross_pay), 2))
print("Medicare Tax: $", round(get_medicare_tax(gross_pay), 2))
print("Federal Tax: $", round(get_federal_tax(gross_pay, withholding_code), 2))
```
