# Lab 2 - Date and Time Functions

This is Lab 2 of Exercise 2.B. In this lab, I am working with datetime functions in Python to work with dates and time.

```python
import datetime
```

# Today

```python
# Creating a variable today that stores the current date and time
today = datetime.datetime.now()
```

```python
# strftime() means string format time
# Using .strftime() to format the date into a readable style.
# %A means full weekday name, %B means full month name, %d means day number, and %Y is for 4-digit year
print("Today is", today.strftime("%A, %B %d, %Y"))
```

# Bumi's birthday

```python
# Creating a variable birhtday to store a specific date
birthday = datetime.date(2023, 5, 1)
```

```python
# Lab wants MM/DD/YYYY, %m is for the month, %d is for the day, %Y is for the 4-digit year
print("Bumi's birthday is", birthday.strftime("%m/%d/%Y"))
```

# Next 90 Days

```python
# Creating a variable ninety_d to store a future date
# Using today the variable created earlier
# timedelta creates the time difference, (days=90) is telling it to move forward by that amount
ninety_d = today + datetime.timedelta(days=90)
```

```python
# strftime formats the future date into the format that the lab wants
# Month day, year. %B for the month, %d for the day and %Y for the 4-digit year
print("90 days from today is", ninety_d.strftime("%B %d, %Y"))
```

# Time for Dinner

```python
# Creating a variable dinner_time to store a specific time
# Putting the hour and minute as 18 for PM because Python uses 24-hour time
dinner_time = datetime.time(18, 15)
```

```python
# %I formats it to 12-hour, %M for minutes and %p for AM or PM
print("Let's meet for dinner at", dinner_time.strftime("%I:%M %p"))
```
